```python
import os
files = [f for f in os.listdir('.') if os.path.isfile(f) and f.endswith('.pdf')]
files
```

```python
from multilingual_pdf2text.pdf2text import PDF2Text
from multilingual_pdf2text.models.document_model.document import Document
import os
from pdf2image import convert_from_path
import img2pdf
import logging
import re

from langdetect import detect_langs 

import re
try: 
    from PIL import Image 
except ImportError: 
    import Image 
import pytesseract 
logging.basicConfig(level=logging.INFO)
pytesseract.pytesseract.tesseract_cmd = 'D:/New folder/tesseract.exe'  # your path may be different

dict1={}
id=0
for i in files:
    id=id+1
    dict1['id']=id
    pdf_document = Document(
        document_path=i,
        language='fas+eng'
        )
    pdf2text = PDF2Text(document=pdf_document)
    content = pdf2text.extract()
    print(content)
    print('**************')
    #     print('type content: ' , type(content))
    #     print('type content[0]: ' , content[2])
    print('len of content: ' , len(content))
    for i in range(len(content)):
        dic = content[i]
        text = dic.get('text')
    #         print('type of text: ', type(text))
        a = 'استاد راهنما'
        if a in text:
    #         print('found: ', i)
            result = re.search(a+'\s.*\n', text)
            print('result: ', result)
            r_strip = result.group(0).strip()
            print('r_strip: ', r_strip)
    #         print('**************')
            s = re.split('\n', r_strip)
            print('s: ', s)
            supervisor = s[len(s)-1]
            print('supervisor : ', supervisor)
            x1 = supervisor.split(' ')
            print(x1)
            print('FistName: ' , x1[0])
            print("LastName: ", x1[1])
            dict1['supervisor_FistName']=x1[0]
            dict1['supervisor_LastName']=x1[1]
    for i in range(len(content)):
            dic = content[i]
            text = dic.get('text')
        #         print('type of text: ', type(text))
            a = 'نگارش'
            if a in text:
        #         print('found: ', i)
                result = re.search(a+'\n\s.*\n', text)
        #         print('result: ', result)
                r_strip = result.group(0).strip()
        #         print('r_strip: ', r_strip)
        #         print('**************')
                s = re.split('\n', r_strip)
        #         print('s: ', s)
                creator = s[len(s)-1]
                print('creator : ', creator)
                x = creator.split(' ')
        #         print(x)
                print('FistName: ' , x[0])
                print("LastName: ", x[1])
                dict1['student_FistName']=x[0]
                dict1['student_LastName']=x[1]
    
    for i in range(len(content)):
                dic = content[i]
                text = dic.get('text')
            #         print('type of text: ', type(text))
                a = 'عنوان پایان نامه'
                if a in text:
            #         print('found: ', i)
                    result = re.search(a+'\s.*\n', text)
            #         print('result: ', result)
                    r_strip = result.group(0).strip()
            #         print('r_strip: ', r_strip)
            #         print('**************')
                    s = re.split('\n', r_strip)
            #         print('s: ', s)
                    title = s[len(s)-1]
                    print('title : ', title)
                    dict1['title']=title
    for i in range(len(content)):
                    dic = content[i]
                    text = dic.get('text')
                #         print('type of text: ', type(text))
                    a = 'کلمات کلیدی'
                    if a in text:
                #         print('found: ', i)
                        result = re.search(':\s.*\n', text)
                #         print('result: ', result)
                        r_strip = result.group(0).strip()
                #         print('r_strip: ', r_strip)
                #         print('**************')
                        s = re.split('\n', r_strip)
                #         print('s: ', s)
                        keyWords = s[len(s)-1]
                        print('key words : ', keyWords)
                        dict1['key words']=keyWords                 
    EnAbstract=None
    # for i in range(1, 5):
    #     dic = content[len(content) - i]
    #     text = dic.get('text')
    #         print('type of text: ', type(text))
    a = 'Abstract'
    if a in text:
    #     print('found: ', len(content) - i)
        start = text.find("Abstract") + len("Abstract")
        end = text.find("Key")
        result = text[start:end]
    #     print('result: ', result)
        r_strip = result.strip()
    #         print('r_strip: ', r_strip)
    #         print('**************')
        s = re.split('\n', r_strip)
    #         print('s: ', s)
        EnAbstract = ''
        for t in s:
            EnAbstract += ' '+t
        EnAbstract = EnAbstract.strip()
        print('Abstract:', EnAbstract)
        dict1['Abstract']=EnAbstract
    RE_USELESS = r'[^\w]'  # remove useless characters
    RE_DIGIT = r"^\d+\s|\s\d+\s|\s\d+$"  # remove digits
    RE_SPACE = r'\s+'  # remove space    
    heading=None
    # for i in range(1, 5):
    #     dic = content[len(content) - i]
    #     text = dic.get('text')
    #         print('type of text: ', type(text))
    a = 'فهرست مطالب'
    if a in text:
    #     print('found: ', len(content) - i)
        start = text.find("فهرست مطالب") + len("فهرست مطالب")
        end = text.find("\u202bفهرست اشکال\u202c")
        result = text[start:end]
    #     print('result: ', result)
        r_strip = result.strip()
    #         print('r_strip: ', r_strip)
    #         print('**************')
        s = r_strip.split('.')
    #     s = [re.sub(r'[0-9]','', str(x)) for x in s]
        s = [re.sub(r'صفحه','',re.sub(r'فهرست مطالب','', (x))) for x in s]

        s = [re.sub(RE_SPACE, r' ', re.sub(RE_USELESS, r' ', re.sub(r'[0-9]','', str(x)))).strip() for x in s]
        while(True):
            try:
                s.remove('')
            except:
                break
        while(True):
            try:
                s.remove(' ')
            except:
                break

    #         print('s: ', s)
        heading = s[:-1]
    #     heading = heading.strip()
        print('Heading:', heading)
```

```python
# footNotes extractor
pytesseract.pytesseract.tesseract_cmd = 'D:/New folder/tesseract.exe'  # your path may be different

footNotes=[{"pageNo", "footNote"}]
for pdf_file in os.listdir(r"C:\Users\taban6"):

    if pdf_file.endswith(".pdf"):
        print(pdf_file)
        pages = convert_from_path(pdf_file, 500 ,poppler_path=r'C:\Program Files (x86)\poppler-0.68.0\bin')
        for i in range(len(pages)):
        # Save pages as images in the pdf
            image=pdf_file.split('.')[0]+ str(i) +'.jpg'
            pages[i].save(image,'JPEG')
            img=Image.open(image)
            left = 0
            top = 4500
            right = 4200
            bottom = 5500
            img_res = img.crop((left, top, right, bottom))
            outfile='TestFotenote.jpg'
            img_res.save(outfile,'JPEG')
            outPdf="C:/Users/taban6/New folder/output.pdf"
            with open(outPdf, "wb") as f:
                f.write(img2pdf.convert(outfile))
## create document for extraction with configurations
            pdf_document = Document(
            document_path="C:/Users/taban6/New folder/output.pdf",
            language='fas+eng')
            
            pdf2text = PDF2Text(document=pdf_document)
            content = pdf2text.extract()
#             print(content)
            a_string =content[0].get('text')
            pageNo=i+1
            matches = ["۱","۲","1","2","3","4","5","6","7","8","9","0"]
            Prohibited_chars=["أ","ب","د","ه","ج","و","ز","ح","ط","ی"]

            match = next((x for x in matches if x in a_string), False)
            Prohibited = next((x for x in Prohibited_chars if x in a_string), False)
#             print(Prohibited)
            if(match!=False and Prohibited==False):
                arayFoot1=a_string[a_string.index(match):].split('\n')
#                 print(arayFoot1)
                new_list=[x for x in arayFoot1 if len(x)>=4]
#                 print(new_list)
                footNotes.append({"pageNo":pageNo,"footNote":new_list})
#             print('footnote is:', footNotes)
print('**************')
footNotes.remove({'pageNo', 'footNote'})
footNote=''
for f in footNotes:
    footNote+="--".join(str(x) for x in f.get('footNote'))    
    f['footNote']=footNote
dict1['footnotes']=footNotes
#     print('type content: ' , type(content))
#     print('type content[0]: ' , content[2])
# print('len of content: ' , len(content))
```
